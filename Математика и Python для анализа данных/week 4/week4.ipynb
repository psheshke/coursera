```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

Парето распределение https://ru.wikipedia.org/wiki/%D0%A0%D0%B0%D1%81%D0%BF%D1%80%D0%B5%D0%B4%D0%B5%D0%BB%D0%B5%D0%BD%D0%B8%D0%B5_%D0%9F%D0%B0%D1%80%D0%B5%D1%82%D0%BE

```python
n = 1000
k = 3.
```

```python
pareto = sts.pareto(b = k)
sample = pareto.rvs(n)

plt.hist(sample, bins = 10, normed=True, label='Гистограмма распределения выборки')

x = np.linspace(1,5,1000)
pdf = pareto.pdf(x)
plt.plot(x, pdf, label='Теоретическое Парето распределение', alpha=5.5)

plt.ylabel('Доля наблюдений')
plt.legend()
plt.xlabel('$x$')
```

\begin{align}
Математическое \; ожидание \; (E): \frac{kx_m}{k-1}, \; если \; k > 1 \\
Дисперсия \; (D): \left ( \frac{x_m}{k-1} \right )^2 \frac{k}{k-2}, \; при \; k > 2
\end{align}

```python
x_m = 1

E = k * x_m / (k - 1)
D = (x_m / (k -1)) **2 * k / (k - 2)
```

```python
print(f'Теоретические значения: E = {E}, D = {D}')
```

```python
sampleE = np.mean(sample)
sampleD = np.std(sample) ** 2
```

```python
pareto.mean(), pareto.var()
```

```python
print(f'Наблюдаемые значения по сгенерированной выборке: E = {sampleE}, D = {sampleD}')
```

Сгенерируем 1000 выборок по n наблюдений и построим распределение их выборочных средних, а также теоретическое нормальное распределение с параметрами, полученными на основании теоретических значений E и D

```python
n = 3
e = []

for i in range(1000):
    sample_n = pareto.rvs(n)
    e.append(np.mean(sample_n))

fig, ax1 = plt.subplots()

ax1.hist(e, normed=True, label='Распределение выборочных средних')
ax1.set_ylim(bottom = 0)
ax1.set_xlim(0, 3)

plt.ylabel('Доля наблюдений')

ax2 = ax1.twinx()

norm_rv = sts.norm(E, D/n)
x = np.linspace(0,3,100)
pdf = norm_rv.pdf(x)

ax2.plot(x, pdf, label='Теоретическое распределение', alpha=0.5, color='red')
ax2.set_ylim(bottom = 0)

plt.ylabel('Аппроксимация')

fig.legend()
```
