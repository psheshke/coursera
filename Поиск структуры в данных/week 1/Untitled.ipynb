```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('checkins.dat',sep='|',low_memory=False, skipinitialspace=True).dropna()
```

```python
df
```

```python
df.columns
```

```python
from sklearn.cluster import MeanShift
```

```python
clustering = MeanShift(bandwidth=0.1).fit(df[['latitude      ', 'longitude     ']].values[:100000,:])

clustering.labels_
```

```python
clustering.cluster_centers_.shape
```

```python
len([n for n in clustering.labels_ if n == 0])
```

```python
def cx(x):
    print(x.id)
    return len([n for n in clustering.labels_ if n == x.id])
```

```python
clust_centers = pd.DataFrame(clustering.cluster_centers_)
clust_centers['id'] = clust_centers.index
```

```python
val, counts = np.unique(clustering.labels_, return_counts=True)
```

```python
clust_centers['count'] = counts
```

```python
clust_centers[clust_centers['count'] > 15].iloc[:, :2]
```

```python
offices = np.array([[33.751277, -118.188740],
[25.867736, -80.324116],
[51.503016, -0.075479],
[52.378894, 4.885084],
[39.366487, 117.036146],
[-33.868457, 151.205134]])
```

```python
clustering.predict(offices)
```

```python
from sklearn.metrics.pairwise import euclidean_distances
```

```python
dist_matrix = pd.DataFrame(euclidean_distances(clust_centers[clust_centers['count'] > 15].iloc[:, :2], offices))
```

```python
dist_matrix['id'] = clust_centers[clust_centers['count'] > 15].id.values
```
