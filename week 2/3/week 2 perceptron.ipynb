```python
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold, cross_val_score
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.preprocessing import scale, StandardScaler
from sklearn.datasets import load_boston
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score
```

```python
train = pd.read_csv('perceptron-train.csv', header=None)

train
```

```python
test = pd.read_csv('perceptron-test.csv', header=None)

test
```

```python
X_train = train.loc[:,1:]

X_train
```

```python
y_train = train[0]

y_train
```

```python
X_test = test.loc[:,1:]

X_test
```

```python
y_test = test[0]

y_test
```

```python
clf = Perceptron(random_state=241)

clf.fit(X_train, y_train)

predictions = clf.predict(X_test)

predictions
```

```python
a1 = accuracy_score(y_test, predictions)

a1
```

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.transform(X_test)
```

```python
#clf = Perceptron(random_state=241)

clf.fit(X_train_scaled, y_train)

predictions_scaled = clf.predict(X_test_scaled)

predictions_scaled
```

```python
a2 = accuracy_score(y_test, predictions_scaled)

a2
```
